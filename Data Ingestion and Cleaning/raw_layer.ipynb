```python
datalake = {
            'csv':['datalake/raw/world_development_indicators/date=20220922/WDIData.csv'],
            'json':['datalake/raw/co2_passenger_cars_emissions/year=2017/co2_emissions_passenger_cars_2017.json',
                    'datalake/raw/co2_passenger_cars_emissions/year=2018/co2_emissions_passenger_cars_2018.json',
                    'datalake/raw/co2_passenger_cars_emissions/year=2019/co2_emissions_passenger_cars_2019.json']
            }
world_dev_indi = spark.read.options(header=True, inferSchema=True).csv(','.join(i for i in datalake['csv']))
# drop one null header column
# world_dev_indi = world_dev_indi.drop("_c66")
co2_cars_emissions = spark.read.json(datalake['json'])
```

```python
print(f"{'--'*10}world_development_indicators{'--'*10}")

# print schema
print('Schema:')
world_dev_indi.printSchema()

# print number of records
print("Number of records: {}".format(world_dev_indi.count()))

# print basic statistics about each column
print("Basic Statistics:")
world_dev_indi.summary().show()

# Show 5 sample rows of data
print("Sample Data")
world_dev_indi.show(n=5)
```

```python
print(f"{'--'*10}co2_emissions_passenger_cars{'--'*10}")

# print schema
print('Schema:')
co2_cars_emissions.printSchema()

# print number of records
print("Number of records: {}".format(co2_cars_emissions.count()))

# print basic statistics about each column
print("Basic Statistics:")
co2_cars_emissions.summary().show()

# Show 5 sample rows of data
print("Sample Data")
co2_cars_emissions.show(n=5)
```
